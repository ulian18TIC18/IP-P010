```python
from faker import Faker
import numpy as np
faker = Faker('pt_BR')
import pandas as pd
import random
from random import uniform
```

```python
num_linhas = 1000
```

```python
df_faker = pd.DataFrame({'SEXO': ['Masculino' if faker.random_int(0, 1) == 0 else 'Feminino' for i in range(num_linhas)]})
```

```python
df_faker['NOME'] = [faker.name_male() if sexo == 'Masculino' else faker.name_female() for sexo in df_faker['SEXO']]
```

```python
df_faker = df_faker[['NOME', 'SEXO']]
```

```python
df_faker['CPF'] = [faker.cpf() for i in range(num_linhas)]
```

```python
df_faker['IDADE'] = [random.randint(18, 28) for i in range(num_linhas)]
```

```python
df_faker['ENDEREÇO DE EMAIL'] = [faker.email() for i in range(num_linhas)]
```

```python
df_faker['NOTA DO ENEM'] = [random.randint(690, 800) for i in range(num_linhas)]
```

```python
df_faker['ABANDONO'] = ['Sim' if faker.random_int(0, 1) == 0 else np.nan for i in range(num_linhas)]
```

```python
df_faker['SEMESTRE DE ABANDONO'] = [random.randint(1, 8) if df_faker.loc[i]['ABANDONO'] == 'Sim' 
                                    else 0 for i in range(num_linhas)]
```

```python
df_faker['CRA APÓS 2º SEMESTRE'] = [uniform(5.0, 10.0) if semestre_abandono >= 2 and semestre_abandono <= 4 else np.nan for semestre_abandono in df_faker['SEMESTRE DE ABANDONO']]
```

```python
df_faker['CRA APÓS 4º SEMESTRE'] = [uniform(5.0, 10.0) if semestre_abandono > 4 and semestre_abandono <= 6 else np.nan for semestre_abandono in df_faker['SEMESTRE DE ABANDONO']]
```

```python
df_faker['CRA APÓS 6º SEMESTRE'] = [uniform(5.0, 10.0) if semestre_abandono > 6 else np.nan for semestre_abandono in df_faker['SEMESTRE DE ABANDONO']]
```
